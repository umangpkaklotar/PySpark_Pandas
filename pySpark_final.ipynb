```python
from pyspark.sql import SparkSession
```

```python
# create the app
spark = SparkSession.builder\
        .appName('pyspark Fundamentaks') \
        .getOrCreate()
```

```python
spark
```

```python
data = [
    ("Zaid",66,15,"india"),
    ("shivarj",78,15,"india"),
    ("john",66,50,"UK"),
    ("sara",97,89,"USA")
]
columns = ["name","age","Experience","country"]
```

```python
df = spark.createDataFrame(data,columns)
```

```python
df.show()
```

```python
df.printSchema()
```

```python
df.select("name","age").show()
```

```python
df.filter(df.age > 66).show()  # use filter by age filtaring
```

```python
df.groupBy("country").count().show()
```
